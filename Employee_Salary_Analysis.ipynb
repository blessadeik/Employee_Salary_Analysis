```python
# Import Libraries
import pandas as pd
```

The employees.csv file contains information about employees of a company with following information: First Name', 'Gender', 'Start Date', 'Last Login Time', 'Salary', 'Bonus %', 'Senior Management', 'Team'

```python
# Read the csv file and provide summary of the DataFrame by printing the object.
employees = pd.read_csv('employees.csv')
# print(employees)

employees.info()
```

```python
#  Fill all the null values in Gender column with “No Gender” and 'Senior Management' column with “Don’t Know”.
employees['Gender'].fillna('No Gender', inplace=True)
employees['Senior Management'].fillna('Dont Know', inplace=True)
display(employees)
```

```python
#  Using the fillna (0), fillna () with method ‘ffill’ and interpolate () function to fill the missing values in 'Salary', 'Bonus %' columns and calculate the average (mean) salary and bonus for each method.

# fillna(0)
employees['Salary'].fillna(0, inplace=True)
employees['Bonus %'].fillna(0, inplace=True)

# fillna()
employees['Salary'].fillna(method='ffill')
employees['Bonus %'].fillna(method='ffill')

# interpolate
employees['Bonus %'].interpolate()
employees['Bonus %'].interpolate()

mean_salary = employees['Salary'].mean()
print('The average salary is:', mean_salary)

mean_bonus = employees['Bonus %'].mean()
print('The average bonus is:', mean_bonus)
```

```python
#  Filter the rows to only show employee with Salaries >100K.
salary_over_100k = employees[employees['Salary'] > 100000]
display(salary_over_100k)
```

```python
#  Print the names of male and female employees with highest and lowest salaries.

# Female min
female_salary = employees[employees['Gender']== 'Female'].copy()
female_salary[female_salary['Salary'] == female_salary['Salary'].min()]
```

```python
# Female max
female_salary = employees[employees['Gender']== 'Female'].copy()
female_salary[female_salary['Salary'] == female_salary['Salary'].max()]
```

```python
# Male max
male_salary = employees[employees['Gender']== 'Male'].copy()
male_salary[male_salary['Salary'] == male_salary['Salary'].max()]
```

```python
# Male min
male_salary = employees[employees['Gender']== 'Male'].copy()
male_salary[male_salary['Salary'] == male_salary['Salary'].min()]
```

```python
#  For each Team, calculate the minimum and maximum salaries.
team_min_max_salary = employees.groupby('Team')['Salary'].agg(['min', 'max']).copy()
print(team_min_max_salary)
```

```python
#  Find the maximum and average salary of Male and Female as different groups.

team_min_max_teamgen = employees.groupby(['Team', 'Gender']).agg({'Salary':['min', 'max']}).copy()
display(team_min_max_teamgen)
```

```python
#  For each combination of Team and Gender, calculate the mean salary.
team_gender_mean_salary = employees.groupby(['Team', 'Gender'])['Salary'].mean().copy()
print(team_gender_mean_salary)
```

```python
print('Number of rows and columns:', employees.shape)
```

```python
#  Check the number of rows and columns and drop those row if 'any' values are missing in a row.
employees.dropna(axis=0, how='any', inplace=True)
print('Number of rows and columns:', employees.shape)
```

```python
#  Find all the employees whose bonuses are in the range of >=5% and <10%.
bonus_range_5_10 = employees[(employees['Bonus %'] >= 5) & (employees['Bonus %'] < 10)]
display(bonus_range_5_10)
```

```python
#  Find the total salary of each employee by considering both base salary and bonus and add it as an extra column “Total Salary” at the end.
employees['Total Salary'] = employees['Salary'] + (employees['Bonus %'] * employees['Salary'] / 100)
display(employees)
```

```python
#  Sort the dataset with 'Start Date' and replace it as the index.
employees['Start Date'] = pd.to_datetime(employees['Start Date'])

employees.sort_values(by='Start Date', inplace=True)
employees.set_index('Start Date', inplace=True)
display(employees)
```

```python
#  Find the employees who joined in the years between 2000-2010.
employees['2000':'2010']
```

```python
#  Use apply() to find the difference between Max and Min Salary and Max and Min Bonus.
salary_min_max_difference = employees[['Salary', 'Bonus %']].apply(lambda x: x.max() - x.min())
print(salary_min_max_difference)
```
